```python
import math
def my_range(start, end, step):
    while start <= end:
        yield start
        start += step
```

```python
def lagrangePolynomial(x_Arr, y_Arr, xx):
   # print(x_Arr, y_Arr)
    l = 0.0
    counter_i = 0
    for i in range(len(x_Arr)):
        q = 1.0
        counter_j = 0
        for j in range(len(x_Arr)):  
            if j != i: 
                q=q*(xx - x_Arr[counter_j]) / (x_Arr[counter_i] - x_Arr[counter_j])
            counter_j = counter_j + 1
        #print(str(q) +'   ' + str(l))
        l = l + y_Arr[counter_i] * q
        counter_i = counter_i + 1
    return l;
```

```python
def func_18th_var(start, end, step):
    x = start 
    #list_arr = [0 for i in range(0, step)]
    y_array = []
    counter_i = 0
    for i in my_range(start, end, step):
        y_array.append(math.sqrt(x) - x)
        #print("%.1f"%(i) + ':'+ str(y_arr[counter_i]))
        x = x + step
        counter_i = counter_i + 1
    return y_array
```

```python
def func_1st_var(start, end, step):
    x = start 
    y_array1 = []
    counter_i = 0
    for i in my_range(start, end, step):
        y_array1.append(math.sin(x)*x)
       # print("%.1f"%(i) + ' : '+ str(y_array1[counter_i]))
        x = x + step
        counter_i = counter_i + 1
    return y_array1
```
