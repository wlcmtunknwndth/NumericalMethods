```python
import math
from prettytable import PrettyTable

def ode_22(x, y):
    return 2*y * math.cos(2*x)

def ode_1(x, y):
    return x*y

def func_1(x):
    return math.e**(0.5*x**2)
```

```python
def my_range(start, end, step):
    while start <= end: 
        yield start
        start +=step
```

```python
def eulers_method(x0, xn, n, y0):
    dx = (xn - x0)/n
    y = y0
    ans = []
    for x in my_range(x0, xn, dx):
        y += dx*ode_1(x, y)
        ans.append(y)
    return ans
```

```python
x0 = 0
xn = math.pi/4
n = 32
h = (xn-x0)/n
y0 = 1 


kk = eulers_method(x0, xn, n,y0)
for i in range(len(kk)):
    error = abs(func_1(x0 + i*h) - kk[i])  
    print(round(x0+i*h, 2), kk[i], func_1(x0 + i*h), error)
```
